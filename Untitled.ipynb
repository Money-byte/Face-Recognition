```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("C:\\Users\\Aprajit Sharma\\Desktop\\Data Vision\\Dog.jpg")
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
(hgt,wth)=image_rgb.shape[:2]
res_image = cv2.resize(image_rgb, (int(wth/2),int(hgt/2)),interpolation = cv2.INTER_CUBIC)
plt.imshow(res_image)
```

```python
flip_image = cv2.flip(image_rgb,0)
plt.imshow(flip_image)
```

```python
flip_img = cv2.flip(image,1)
plt.imshow(flip_img)
```

```python
edges = cv2.Canny(flip_image,100,200)
plt.imshow(edges)
```

```python
Gaussian_img = cv2.GaussianBlur(image_rgb,(7,7),0)
cv2.imshow('Gaussian Blur',Gaussian_img)
cv2.waitKey(0)
```

```python
median = cv2.medianBlur(image,5)
cv2.imshow('Median Blurring',median)
cv2.waitKey(0)
```

```python
bilateral = cv2.bilateralFilter(image,9,75,75)
cv2.imshow('Bilateral Blurring ',bilateral)
cv2.waitKey(0)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace



```

```python
img=cv2.imread("C:\\Users\\Aprajit Sharma\\Desktop\\Data Vision\\Happy.jpg")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
!python -v
```

```python
!python --version
```

```python
predictions=DeepFace.analyze(img)
```

```python
print(predictions)
```

```python
print(predictions['dominant_emotion'])
```

```python
facecascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = facecascade.detectMultiScale(gray, 1.1, 4)

for(x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 4)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,predictions['dominant_emotion'],(60,90),font,4,(0,280,0),2,cv2.LINE_4)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
